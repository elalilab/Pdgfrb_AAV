# Visualizing cell numbers by heatmaps

In this short notebook, we use the `heatmap` library from BrainGlobe [https://github.com/brainglobe/brainglobe-heatmap] to visualize the number of cells in specific brain regions.

## Load libraries

```python
import brainglobe_heatmap as bgh
```

## Setup regions to visualize

We create the object `values` to store the regions and corresponding cell number to be ploted. Please note that this takes as an input the acronyms for the mouse Allen brain atlas that can be reviewd in the `Atlas_Regions.csv` file available in the GitHub repository [https://github.com/elalilab/Pdgfrb_AAV].

```python
import brainglobe_heatmap as bgh

values = dict(  # scalar values for each region
   CTX=60,
    CNU=28,
    HPF=44,
    TH=10,
    MB=12,
    HY=18,       
)
```

## Generate the visualization

Here, we feed the values and setup other parameters. Importantly, the `position` parameter defines the brain level to be ploted. Higher values bring the plot to forebrain regions. We plot the same values in two different levels, at 6000 and 7200.

```python
f = bgh.heatmap(
    values,
    position=(6000),  # when using a named orientation you can pass a single value!
    orientation="frontal",  # 'frontal' or 'sagittal', or 'horizontal' or a tuple (x,y,z)
    title="",
    vmin=0,
    vmax=60,
    atlas_name = "allen_mouse_25um",
    cmap='viridis',
    format="2D",
)
f.show(xlabel='AP (m)', ylabel='DV (m)')
```

```python
f = bgh.heatmap(
    values,
    position=(7500),  # when using a named orientation you can pass a single value!
    orientation="frontal",  # 'frontal' or 'sagittal', or 'horizontal' or a tuple (x,y,z)
    title="",
    vmin=0,
    vmax=60,
    atlas_name = "allen_mouse_25um",
    cmap='viridis',
    format="2D",
)
f.show(xlabel='AP (m)', ylabel='DV (m)')
```
